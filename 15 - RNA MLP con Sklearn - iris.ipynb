## Redes neuronales artificiales: Perceptrón multicapa

```python
# Cargar módulos
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import pandas as pd
```

```python
data = sns.load_dataset("iris")
data.head()
```

```python
train, test = train_test_split(data, test_size = 0.4, stratify = data["species"], random_state = 31)
```

```python
X_train = train[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
Y_train = train["species"]
X_test = test[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
Y_test = test["species"]
```

```python
#Especificar caracteristicas, modelo estocástico = "sgd"
mlp = MLPClassifier(hidden_layer_sizes = (3,3), solver = "sgd", learning_rate_init = 0.01, max_iter=500)
```

```python
# Codificar la variable, que es la variable clasificatoria
data["species"].unique()
```

```python
#Codificar la variable de respuesta [categórica]
from sklearn.preprocessing import LabelEncoder
codificador = LabelEncoder()
```

```python
data["species"] = codificador.fit_transform(data["species"])
species = pd.DataFrame({"species": ["setosa", "versicolor", "virginica"]})
species
```

```python
data.iloc[70:80]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
#Crear copia por seguridad
noescalado = X_train.copy()
X_train_array = sc.fit_transform(X_train.values)
```

```python
#Se coloca como argumentos los datos transformados, el indice del train anterior, al igual que las columnas
X_train = pd.DataFrame(X_train_array, index = X_train.index, columns = X_train.columns)
```

```python
X_test_array = sc.fit_transform(X_test.values)
X_test = pd.DataFrame(X_test_array, index = X_test.index, columns = X_test.columns)
```

```python
X_test
```

```python
mlp.fit(X_train, Y_train)
```

```python
mlp.coefs_
```

```python
from sklearn.metrics import accuracy_score
```

```python
mlp.score(X_train, Y_train)
```

```python
prediccion_train = mlp.predict(X_train)
prediccion_test = mlp.predict(X_test)
```

```python
train_punteo = accuracy_score(prediccion_train, Y_train)
test_punte = accuracy_score(prediccion_test, Y_test)
```
