```python
import numpy as np
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
import pandas as pd
```

```python
data = sns.load_dataset("iris")
data.head(10)
```

y = respuesta
x = variable explicativa
m = pendiente (razon de cambio)
b = interpecto

```python
data["species"].unique()
```

```python
data["species"].value_counts()
```

```python
x = data["petal_length"]
y = data["petal_width"]
```

```python
# Agregar columna constante = 1 para cálculo de intercepto
x = sm.add_constant(x)
```

```python
# Y en función de X
resultado = sm.OLS(y,x).fit()
```

```python
resultado.summary()
```

```python
# R de pearson (correlacion)= Que tanta relación hay entre dos variables. 0 = no relacion, 0.2= muy baja, 0.4 = baja,  1 = positiva, -1 = total
# R2 (coeficiente de determinación)
resultado.params
```

```python
resultado.rsquared
```

```python
import math
```

```python
math.sqrt(resultado.rsquared)
```
